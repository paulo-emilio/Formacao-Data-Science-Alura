## 07_02 - Criando Novas Variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Criando a variável 'Valor Bruto'
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
# Criando a variável 'Valor m2'
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
# Arredondando para duas casas decimais
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

```python
# Criando a variável 'Valor Bruto m2'
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
# Criando a variável 'Tipo Agregado'
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
# apply -> Permite uma função seja aplicada a cada registro do DataFrame
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```

## 07_04 - Excluindo Variáveis

```python
# Criando um DataFrame copiando somente essas variáveis
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_aux.head(10)
```

```python
# Ecluindo uma variável com del
del dados_aux['Valor Bruto']
dados_aux
```

```python
# Ecluindo uma variável com pop
dados_aux.pop('Valor Bruto m2')
dados_aux
```

```python
# Ecluindo variáveis com drop (o drop funciona em linhas e colunas)
dados_aux.drop(['Tipo Agregado', 'Valor m2'], axis=1, inplace=True)  # axis=1 define que será excluído nas colunas
dados_aux
```

```python
# Excluindo Valores Brutos do DataFrame original
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis=1, inplace=True)
```

```python
dados.head(10)
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep=';', index = False)
```
