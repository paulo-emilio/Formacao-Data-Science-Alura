## 09_02 - Identificando e Romovendo Outliers

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (14,6)) # figsize = tamnho da figura
```

<img src="imgs/boxplot.jpg">

```python
# Criando um box-lot como na imagem acima 
Q1 = valor.quantile(.25) # Primeiro quartile
Q3 = valor.quantile(.75) # Terceiro quartile
IIQ = Q3 - Q1 # Intervalo inter quartile
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
selecao = (valor >= limite_inferior) & (valor <= limite_superior)
dados_new = dados[selecao]
```

```python
dados_new.boxplot(['Valor'])
```

## 09_02 - Identificando e Romovendo Outliers por Grupo

```python
# Por tipo
dados.boxplot(['Valor'], by=['Tipo'])
```

```python
# Grupo para a variável valor por tipo
grupo_tipo = dados.groupby('Tipo')['Valor']
grupo_tipo.groups
```

```python
Q1 = grupo_tipo.quantile(.25) # Primeiro quartile
Q3 = grupo_tipo.quantile(.75) # Terceiro quartile
IIQ = Q3 - Q1 # Intervalo inter quartile
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
dados_new = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    slecao = eh_tipo & eh_dentro_limite 
    dados_selecao = dados[selecao]
    dados_new = pd.concat(dados_new, dados_selecao)
```

```python
dados_new.boxplot(['Valor'], by=['Tipo'])
```

```python
dados_new.to_csv('dados/aluguel_residencial_sem_outliers.csv', sep=';', index=False)
```
