```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(15,8))
```

```python
dados = pd.read_csv('dados/aluguel.csv', sep=';')
dados.head()
```

```python
area = plt.figure()
```

```python
# Criando um plot de 4 gráficos
g1 = area.add_subplot(2,2,1) # 2 linhas, 2 colunas, e esse vai ficar na posição 1
g2 = area.add_subplot(2,2,2) # posição 2
g3 = area.add_subplot(2,2,3) # posição 3
g4 = area.add_subplot(2,2,4) # posição 4
```

```python
# Criando o gráfico 1
g1.scatter(dados.Valor, dados.Area)  # Scatter Plot
g1.set_title('Valor X Área')
```

```python
area
```

```python
# Criando o gráfico 2
g2.hist(dados.Valor)
g2.set_title('Histograma')
area
```

```python
# Criando o gráfico 3
dados_g3 = dados.Valor.sample(100) # amostra de tamanho 100
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')
area
```

```python
grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
values = grupo.mean().values
g4.bar(label, values)
g4.set_title('Valor Médio por Tipo')
area
```

```python
# Salvando a imagem
area.savefig('dados/grafico.png', dpi=300, bbox_inches='tight') # bbox retira a borda branca em volta
```
