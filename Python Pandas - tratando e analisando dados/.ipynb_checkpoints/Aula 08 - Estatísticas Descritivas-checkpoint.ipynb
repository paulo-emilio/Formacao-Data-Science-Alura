## 07_02 - Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep=';')
dados.head(10)
```

```python
dados['Valor'].mean()
```

```python
# Select dos bairros que vamos fazer a média de preço
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
select = dados['Bairro'].isin(bairros)
dados = dados[select]
dados
```

```python
# Check
dados['Bairro'].unique()
```

```python
# Criando um 'groupby' dos Bairros
grupo_bairro = dados.groupby('Bairro')
```

```python
# O que foi criado
grupo_bairro.groups
```

```python
# Criando um laço para vizualizar o que foi criado
for bairro, dados in grupo_bairro:
    print(bairro)
    print('--'*36)
    print(dados)
```

```python
# Cada 'dado' de cada bairro, é um DataFrame
for bairro, dados in grupo_bairro:
    print(type(dados))
```

```python
# Calculando a média do Valor de cada Bairro
for bairro, data in grupo_bairro:
    media = data['Valor'].mean()
    print(f'{bairro}__{media:.2f}')
```

```python
# Calculando a média do Valor de cada Bairro mais facilmente
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## 08_04 - Estatísticas Descritivas

```python
# Frequência - Média - Desvio Padrão - Valor Mínimo- 25% - 50% - 75% - Valor Máximo
grupo_bairro['Valor'].describe().round(2)
# Percebemos que Botafogo possui um imóvel com valor de locação 4500000.0, ou seja, está errado,
# também um errado em Barra da Tijuca
```

```python
# .aggregate() e reanme()
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns={'min': 'Mínimo', 'max': 'Máximo'})
```

### matplotlib  (melhor verificação dos dados)

```python
%matplotlib inline  # configuração para gráficos
import matplotlib.pyplot as plt 
plt.rc('figure', figsize = (20, 10))
```

```python
# Exibindo um gráfico com o desvio padrão (std) para verificar melhor
fig = grupo_bairro['Valor'].std().plot.bar()
fig
```

```python
# Colocando nomes dos eixos e Título
fig = grupo_bairro['Valor'].std().plot.bar()
fig.set_ylabel('Valor do Aluguel', {'fontsize':14})
fig.set_xlabel('Bairro', {'fontsize':14})
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```
