## 04_02 Visualizando Gráficos por Categoria

```python
import pandas as pd
tmdb = pd.read_csv('dados/tmdb_5000_movies.csv')
tmdb.head()
```

```python
# Series de 'original_language'
tmdb['original_language'].value_counts()
```

```python
# index
tmdb['original_language'].value_counts().index
```

```python
# values
tmdb['original_language'].value_counts().values
```

```python
# tansformando Series em DataFrame
tmdb['original_language'].value_counts().to_frame()
```

```python
# colocando index numeral
contagem_de_lingua = tmdb['original_language'].value_counts().to_frame().reset_index()
contagem_de_lingua.columns = ['original_language', 'total']
contagem_de_lingua.head()
```

```python
import seaborn as sns
```

```python
# plotando com o seaborn 
sns.barplot(x='original_language', y='total', data = contagem_de_lingua)
```

#### Fazendo isso de uma forma mais simples com o seaborn

```python
# obs: se a versão do seaborn estiver antiga, usar:
'''!pip install seaborn==0.9.0'''
```

```python
sns.catplot(x='original_language', kind='count', data=tmdb)
# faça um plot usando 'original_language' do tipo 'count' a partir do tmdb
# podemos observar que plot sem ordem
```

## 04_03 - Passando uma mensagem através de vizualizações

```python
# Gráfico de Pizza/Torta (pie)
import matplotlib.pyplot as plt
plt.pie(contagem_de_lingua['total'], labels=contagem_de_lingua['original_language'])
# KKKKKKKKKKKKK
```

```python
# Seprando os dados para criar uma visualização 'en' vs 'outros'
tot_por_lingua = tmdb['original_language'].value_counts()
tot_geral = tot_por_lingua.sum()
tot_ingles = tot_por_lingua.loc['en']
tot_resto = tot_geral - tot_ingles
print(tot_ingles, tot_resto)
```

```python
# Criando o DataFrame com os dados separados
dados = {
    'lingua': ['ingles', 'outros'],
    'total': [tot_ingles, tot_resto]
}
dados = pd.DataFrame(dados)
dados
```

```python
# plotando barra
sns.barplot(x='lingua', y='total', data=dados)
```

```python
# plotrando torta
plt.pie(dados['total'], labels=dados['lingua'])
```
